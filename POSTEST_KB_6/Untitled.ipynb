```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets as skdt
import scipy
import matplotlib
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
tf_array = pd.read_csv('transfusion.csv')
type(tf_array)
```

```python
df_tf = pd.read_csv('transfusion.csv')
df_tf.head(5)
```

```python
points = df_tf[['Frequency (times)', 'Time (months)']]
```

```python
plt.scatter(points['Frequency (times)'], points['Time (months)'])
plt.show()
```

```python
model = KMeans(n_clusters=2 )
```

```python
model.fit(points)
```

```python
labels = model.labels_
```

```python
labels
```

```python
centroid = model.cluster_centers_
```

```python
centroid
```

```python
plt.scatter(points['Frequency (times)'], points['Time (months)'], alpha=.5, c=labels)
plt.scatter(centroid[:,0], centroid[:,1], marker='D', s=50)
plt.show()
```

```python
cluster_sum = range(1,5)
inertia =[]

for cluster in cluster_sum:
    model = KMeans(n_clusters=cluster)
    model.fit(points)
    inertia.append(model.inertia_)
    
plt.plot(cluster_sum, inertia, '-o')
plt.xticks(cluster_sum)
plt.show()
```
