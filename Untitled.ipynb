```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1,2,3,4,5,6]
y = [1,4,3,6,2,5]
z = [1,3,4,2,5,3]

plt.figure(figsize = (10, 5))
plt.title("GRAFIK")
plt.xlabel("nilai X")
plt.ylabel("nilai Y")
plt.plot(x, y, label="nilai y", color="darkblue", marker="o")
plt.bar(x, z, label="nilai z", color="skyblue")
plt.legend(loc="lower right")
plt.show()
```

```python
kasus = pd.read_csv("employee.csv")
kasus.head(5)
```

```python
plt.figure(figsize = (5, 3))
plt.plot(kasus["Age"], kasus["DailyRate"], label = "Swab", linestyle="--") 
plt.plot(kasus["Age"], kasus["Education"], label = "negative", linestyle="--")
plt.plot(kasus["Age"], kasus["DistanceFromHome"], label = "positive", linestyle="--")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (5,3))
plt.pie(df["BusinessTravel"].value_counts(), labels=["Travel_Rarely", "Travel_Frequently"], colors=["darkblue", "skyblue"], startangles=90, autopct="%.1f%%", explode=(0.1, 0))
plt.legend(loc="lower right")
plt.show()
```

```python
plt.figure(figsize = (5,3))
plt.pie(df["Attrition"].value_counts(), labels=["Yes", "No"], colors=["darkblue", "skyblue"], explode=(0.1, 0.1, 0))
plt.legend(loc="lower right")
plt.show()
```
